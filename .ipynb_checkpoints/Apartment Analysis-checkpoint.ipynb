```python
import pandas as pd
import matplotlib  as plt
```

```python
area = "Seattle, WA"
path = "./data/"+area
vendor_info_df = pd.read_csv(path + "_vendor_info.csv")
listing_info_df = pd.read_csv(path + "_listing_info.csv")
aptType_info_df = pd.read_csv(path + "_aptType_info.csv")
amenites_info_df =pd.read_csv(path + "_amenites_info.csv")
```

## Display some basic information about rent without doing any dataframe merges yet

```python
print("Basic rent Information")
print("The max rent in the area for : {}".format(area))
print(listing_info_df['rent'].max())
print("The min rent in the area for : {}".format(area))
print(listing_info_df[listing_info_df.rent !=0]['rent'].min())
print("The avg rent in the area for : {}".format(area))
print(listing_info_df[listing_info_df != 0][listing_info_df.sqft > 700][listing_info_df.sqft < 2000]['rent'].mean())
print()
```

```python
vendor_listing_df = pd.merge(vendor_info_df, listing_info_df, left_on="propertyName", right_on="propertyName")
vendor_listing_df[vendor_listing_df != 0][vendor_listing_df.sqft > 700][vendor_listing_df.sqft < 2000][vendor_listing_df.neighborhood =="Belltown"].groupby("bedRoomNumber").count()
```

```python
vendor_listing_df[vendor_listing_df != 0][vendor_listing_df.sqft > 700][vendor_listing_df.sqft < 2000][vendor_listing_df.neighborhood =="First Hill"].groupby("bedRoomNumber").count()
```

```python
amenities = pd.merge(vendor_listing_df, amenites_info_df, left_on="propertyName", right_on="propertyName")
print(amenities)
```

```python
amenitY_groupby = amenities.groupby(["neighborhood"])["amenity"].value_counts(normalize=True)
```

## Display some basic information about sqft without doing any dataframe merges

```python
print("Basic sqft Information")
print("The max sqft in the area for : {}".format(area))
print(listing_info_df['sqft'].max())
print("The min sqft in the area for : {}".format(area))
print(listing_info_df[listing_info_df.sqft !=0]['sqft'].min())
print("The avg sqft in the area for : {}".format(area))
print(listing_info_df[listing_info_df.sqft > 700][listing_info_df.sqft < 2000]['sqft'].mean())
print()
```
